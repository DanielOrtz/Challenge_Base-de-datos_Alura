```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
envios_promedio = [
    envio_promedio_tienda,
    envio_promedio_tienda2,
    envio_promedio_tienda3,
    envio_promedio_tienda4
]

plt.figure()
plt.plot(
    tiendas,
    envios_promedio,
    marker='o',
    linestyle='-'
)

plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de envío promedio')


for i, valor in enumerate(envios_promedio):
    plt.text(i, valor, f'${valor:,.2f}', ha='center', va='bottom')

plt.show()
```

##**INFORME FINAL**


**Introducción**

El objetivo de este análisis fue evaluar el desempeño de cuatro tiendas con base en diferentes factores relacionados con sus ventas y la experiencia de los clientes. Para ello, se analizaron los ingresos totales, las categorías de productos más y menos vendidas, los productos con mayor y menor demanda, las calificaciones promedio otorgadas por los clientes y el costo de envío promedio de cada tienda. A partir de estos datos y de las visualizaciones generadas, se busca determinar en qué tienda le conviene más al Sr. Juan vender sus productos, considerando tanto el potencial de ingresos como la satisfacción del cliente.

**Desarrollo**
1. Análisis de ingresos totales por tienda

En este ejercicio se analizaron los ingresos totales de cada tienda sumando el precio de todos los productos vendidos. Los resultados mostraron que la Tienda 3 obtuvo el mayor ingreso total, seguida por la Tienda 4, mientras que la Tienda 1 y la Tienda 2 presentaron ingresos ligeramente menores. Este resultado indica que la Tienda 3 tiene un mayor volumen de ventas y una mejor capacidad para generar ingresos en comparación con las demás.

2. Análisis de categorías de productos más y menos vendidas

Posteriormente, se analizó la cantidad de ventas por categoría de producto. En todas las tiendas se observó un comportamiento similar, donde la categoría Muebles fue la más vendida. Por ejemplo, en la Tienda 1 se registraron aproximadamente 465 ventas de muebles, mientras que la categoría Electrónicos alcanzó cerca de 448 ventas. En contraste, categorías como Libros e Instrumentos musicales presentaron las menores cantidades de ventas, con valores cercanos a las 170 ventas. Estos resultados muestran una clara preferencia de los clientes por productos de mayor tamaño y uso doméstico.

3. Identificación de productos más y menos vendidos

En este apartado se identificaron los productos individuales más y menos vendidos en cada tienda. Los resultados mostraron que los productos más vendidos pertenecen principalmente a la categoría de Muebles, lo que coincide con el análisis anterior. Por otro lado, los productos menos vendidos se concentraron en categorías como Libros y Instrumentos musicales, lo que indica una menor demanda de estos artículos. Este análisis permite entender con mayor claridad qué productos impulsan las ventas y cuáles tienen un impacto menor.

4. Análisis de calificaciones promedio de los clientes

En cuanto a la satisfacción de los clientes, se calcularon las calificaciones promedio para cada tienda. Los resultados fueron los siguientes: la Tienda 1 obtuvo una calificación promedio de 3.98, la Tienda 2 de 4.04, la Tienda 3 de 4.05 y la Tienda 4 de 4.00. Estos valores indican que todas las tiendas cuentan con una buena aceptación por parte de los clientes, aunque la Tienda 3 destaca ligeramente al presentar la calificación promedio más alta.

5. Análisis del costo de envío promedio por tienda

Finalmente, se analizó el costo de envío promedio de cada tienda. Los resultados mostraron que los costos de envío varían considerablemente entre tiendas, debido al tipo de productos vendidos. La Tienda 3 y la Tienda 4 presentaron los costos de envío promedio más altos, mientras que la Tienda 1 y la Tienda 2 registraron costos ligeramente menores. Esta diferencia se explica principalmente por la venta de productos voluminosos, como muebles, que elevan el costo de envío promedio.

**Conclusión**

Con base en los análisis realizados, se puede concluir que la Tienda 3 es la mejor opción para que el Sr. Juan venda sus productos. Esta tienda destaca principalmente por generar el mayor ingreso total, contar con una alta demanda en las categorías más vendidas, especialmente muebles, y presentar la mejor calificación promedio de los clientes, lo que refleja un buen nivel de satisfacción.

Aunque el costo de envío promedio en la Tienda 3 es más alto en comparación con algunas otras tiendas, este factor es asumido por el cliente, por lo que no representa una desventaja directa para el vendedor. En conjunto, los resultados muestran que la Tienda 3 combina un alto volumen de ventas con una buena percepción por parte de los clientes, lo que la convierte en la alternativa más conveniente para maximizar las oportunidades de venta del Sr. Juan.
